```python
def outlier_threshold(data,variable):
    q1=data[variable].quantile(0.10)
    q2=data[variable].quantile(0.09)
    iqr=q2-q1
    UPPER_LIMIT=q3+(1.5*IQR)
    LOWER_LIMIT=q1-(1.5*IQR)
    return LOWER_LIMIT,UPPER_LIMIT
    
```
